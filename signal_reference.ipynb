```python
import matplotlib.pyplot as  plt
import numpy as np
from scipy.fft import fft
from sklearn.cross_decomposition import CCA
```

```python
target_freq = 12
sr = 250
length = 5*250
reference_signals = []
t = np.arange(0, (length/(sr)), step=1.0/(sr))
#First harmonics/Fundamental freqeuncy
reference_signals.append(np.sin(np.pi*2*target_freq*t))
reference_signals.append(np.cos(np.pi*2*target_freq*t))
#Second harmonics
reference_signals.append(np.sin(np.pi*4*target_freq*t))
reference_signals.append(np.cos(np.pi*4*target_freq*t))
reference_signals = np.array(reference_signals)

squez = np.squeeze(reference_signals)
squez.shape
```

```python
a1 = 1.0
a2 = 0.3

signal = a1*np.sin(np.pi*2*t*target_freq) + a2*np.sin(np.pi*4*t*target_freq)

plt.figure(figsize=(30,10))
plt.plot(t,signal)
plt.show()

```

```python
x = fft(signal)
N = len(x)
n = np.arange(N)
sample_rate = sr
T = N/sample_rate
freq = n/T 
n_oneside = N//2
f_oneside = freq[:n_oneside]
fig, ax = plt.subplots(figsize=(20,10))
ax.plot(f_oneside, np.abs(x[:n_oneside]))
ax.set_xticks(np.arange(min(f_oneside), max(f_oneside)+1, 1.0))
ax.grid()
plt.show()
```
